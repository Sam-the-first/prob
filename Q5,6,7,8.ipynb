```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy
import scipy.stats as stats
import nashpy as nash
```

```python
df = pd.read_csv("array_2d.csv", header= None)
```

```python
df
```

```python
plt.figure(figsize = (20, 20))
beta0 = []
beta1 = []
nrvm = []
nrvs = []
for i in range(6):
    print("company number "+str(i+1)+" Data")
    x=np.arange(1 , 31 , 1)
    y= np.array(df.iloc[:,i])
    expected = np.array([x, y])
    coef=np.polyfit(x,y,1)
    ye=np.poly1d(coef)
    coef = np.array(coef)
    beta0.append(coef[1])
    beta1.append(coef[0])
    
    plt.subplot(6,3 ,i*3+1)
    plt.plot(x,y,".r",x,ye(x),"-y")
    plt.title("company number " + str(i+1))
    plt.savefig("5_"+str(i+1) + ".png")
    print("company number "+ str(i+1)+" beta0 = "+str(round(beta0[i],2)))
    print("company number "+ str(i+1)+" beta1 = "+str(round(beta1[i],2)))
    rv=[]
    for j in range(len(y)):
        rv.append(ye(x[j])-y[j])
#     plt.show()
    plt.subplot(6,3 ,i*3+2)
    
    plt.hist(rv,bins=20)
    plt.title("histogram of difference yhat and y")
    plt.savefig("6_histogram of difference yhat and y number "+str(i+1) + ".png")
#     plt.show()
    nrv=np.array(rv)
    nrvm.append(nrv.mean())
    nrvs.append(nrv.std())

    print("company number "+ str(i+1)+" sigma = "+str(round(nrvs[i],2)))
    normaldist=np.random.normal(nrvm[i],nrvs[i],len(nrv))
    
    plt.subplot(6,3 ,i*3+3)
    plt.hist(normaldist,bins=20)
    plt.title("histogram of normal")
    if i==5:
        plt.savefig("6_histogram of normal number "+str(i+1) + ".png")
#     plt.show()
    print("p-value= " +str(round((stats.ttest_ind(normaldist , nrv).pvalue),4)))
    if(stats.ttest_ind(normaldist , nrv).pvalue < 0.05) :
      print("𝐸𝑣𝑒𝑛𝑡 𝐻1 ∶> No,it isn't like normal distribution!")
    else :
      print ("𝐸𝑣𝑒𝑛𝑡 𝐻0 ∶> Yes,it's like normal distribution!")
    print("-----------------------------------------------")

    plt.tight_layout()
```

```python
Utility1 = np.zeros((6,6))
Utility2 = np.zeros((6,6))
for i in range(1,7):
    for j in range(1,7):
        ai = 7 - i
        bj = np.sqrt(49-j*j)
        Hi = 2.1*(7-(i))
        mi = beta1[i-1]
        Hj = 2.1*(7-(j))
        mj = beta1[j-1]
        if i==j:
            Yi = ai+bj
            Yj = Yi
        else:
            Yi = ai
            Yj = bj
        E1 = ai*mi*(1+Yi/Hi)
        E2 = bj*mj*(1+Yj/Hj)
        Utility1[i-1,j-1] = round(E1,3)
        Utility2[i-1,j-1] = round(E2,3)
print('compony 1 utility'+'\n')
print(Utility1)
print("compony 2 utility"+'\n')
print(Utility2)
print("compony 1 rounded"+"\n")    
print(Utility1.astype(int))
print("compony 2 rounded")
print(Utility2.astype(int))
```
